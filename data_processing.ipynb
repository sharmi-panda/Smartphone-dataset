```python
if "ram" in df.columns:
    df['ram_gb'] = df['ram'].str.extract(r'(\d+)').astype(float)


if "battery" in df.columns:
    df['battery_mah'] = df['battery'].str.extract(r'(\d+)').astype(float)


if "display" in df.columns:
    df['display_inches'] = df['display'].str.extract(r'(\d+\.?\d*)').astype(float)


if "release_date" in df.columns:
    df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
df.info()
```

```python
df.head()
```

```python
df.to_csv("smartphones_cleaned.csv", index=False)
```

```python
def recommend_products(keyword, df, column="model"):
    """
    Recommends products containing the keyword (case-insensitive).
    """
    keyword = keyword.lower()
    results = df[df[column].str.lower().str.contains(keyword, na=False)]

    if results.empty:
        print(f"No products found for: {keyword}")
    else:
        return results
        
recommend_products("OnePlus", df)
```
